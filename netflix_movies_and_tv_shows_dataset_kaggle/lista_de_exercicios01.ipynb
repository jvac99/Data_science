```python
# importações

import pandas as pd 
import numpy as np
```

```python
# Lendo o dataset

uri = "https://raw.githubusercontent.com/jvac99/Data_science/main/netflix_movies_and_tv_shows_dataset_kaggle/netflix_titles.csv"
data = pd.read_csv(uri)
```

```python
# As 5º primeiras linhas ordenadas em ordem decrescente pela coluna release_year

data.sort_values(by=['release_year'], ascending=False).head()
```

```python
# Número de linhas e colunas

tuple_linhas_colunas = data.shape
print("Quantidade de linhas: ", tuple_linhas_colunas[0])
print("\n")
print("Quantidade de colunas: ", tuple_linhas_colunas[1])
```

```python
columns_index = data.columns
print('Quantidade de colunas: ', len(columns_index))
print('\n')
print(columns_index)
```

```python
# Descrição do Dataset

data.describe()
```

```python
# Media, o desvio padrao, e os valores minimo e maximo da serie release_year.

release_year = data['release_year']
print("Média da serie: ", release_year.mean())
print("\n")
print("Mediana da serie: ", release_year.mean())
print("\n")
print("Desvio padrão da serie: ", release_year.std())
print("\n")
print("Nota mínima da serie: ", release_year.min())
print("\n")
print("Nota máxima da serie: ", release_year.max())
```
